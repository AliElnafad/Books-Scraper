```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
import traceback
```

```python
DRIVE_PATH = 'D:\Projects\chromedriver_win32\chromedriver.exe'
driver = webdriver.Chrome(DRIVE_PATH)
```

```python
URL = 'https://books.toscrape.com/index.html'
driver.get(URL)

resultSet = driver.find_element(By.CSS_SELECTOR ,'#default > div > div > div > aside > div.side_categories > ul > li > ul' )
options = resultSet.find_elements(By.TAG_NAME , "li")
#link = options.find_element(By.TAG_NAME , 'a')
```

```python
#books_number = len(books)
prices = []
rate = [] 
titles = []
catogeries = []
availabilities = []
```

```python
print(driver.current_url)
```

```python
for item in range(len(options)):
    #item.click()
    resultSet = driver.find_element(By.CSS_SELECTOR ,'#default > div > div > div > aside > div.side_categories > ul > li > ul' )
    options = resultSet.find_elements(By.TAG_NAME , "li")
    cat = options[item].find_element(By.TAG_NAME , 'a') 
    print(driver.current_url)
    cat.click()
    sleep(2)
    
    while True: #loop for each catogery
        #Scrap all info per page
        books_Per_Page = driver.find_element(By.CSS_SELECTOR ,'#default > div > div > div > div > section > div:nth-child(2) > ol' )
        books = books_Per_Page.find_elements(By.TAG_NAME , "li")
        bookNumber = len(books)
        sleep(1)
        for bookN in range(bookNumber) : #loop for each book in page
            print(len(books))
            print(driver.current_url)
            sleep(1)


            catogery = books[bookN].find_element(By.XPATH, '//*[@id="default"]/div/div/div/div/div[1]/h1'   )
            catogeries.append(catogery.text)
            print("catogery:"  +catogery.text)

            #To Get Title 
            title = books[bookN].find_element(By.TAG_NAME , 'h3')
            titles.append(title.text)
            print("title: "+title.text)

            #To Get Rate
            bookRate = books[bookN].find_element(By.CSS_SELECTOR , '#default > div > div > div > div > section > div:nth-child(2) > ol > li:nth-child('+ str(bookN + 1)  + ') > article > p').get_attribute('class')
            rate.append(bookRate[12:])
            print("Rate: " + bookRate[12:])

            #To get Price and  instock availability
            availability = books[bookN].find_element(By.CSS_SELECTOR , 'div.product_price')
            product_detials = (availability.text).split('\n')
            price = product_detials[0]
            availability = product_detials[1]
            
            prices.append(price)
            availabilities.append(availability)
            
            print('price '+ price)
            print('availability '+ availability)
            print('---'*70)

        if bookNumber >= 20 :
            #To Next Page With The Same Gatogery
            sleep(2)
            div_btn = books[bookN].find_element(By.XPATH ,'//*[@id="default"]/div/div/div/div/section/div[2]/div')
            btn = div_btn.find_element(By.LINK_TEXT , 'next')
            btn.click()
            
            sleep(3)


        else:
            break
        print("Scraper passed from here ( page level ) ")  
        
    #prin 0 ttt')
    print("Scraper passed from here( catogery level )")
    sleep(2) 
        
    
  
    '''#To Get Price
    price = books[bookN].find_element(By.CLASS_NAME , 'price_color')
    prices.append(price.text)'''
```

```python
data = [titles , catogeries , rate   , prices  , availabilities ]
```

```python
#Are all books scraped ?
for item in data:
    print(len(item))
```
